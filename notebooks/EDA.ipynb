```python
#Create visualizations of spending

# Pivot the DataFrame to have countries as columns and years as index
pivot_df = df.pivot(index='Year', columns='Country', values='Spending (2020 USD)')

# Create the line chart
plt.figure(figsize=(12, 8))
for country in pivot_df.columns:
    plt.plot(pivot_df.index, pivot_df[country], label=country)

# Set labels and title
plt.xlabel('Year')
plt.ylabel('Spending (2020 USD)')
plt.title('Military Expenditure for Each Country Over the Years')
plt.legend(loc='upper left', bbox_to_anchor=(1, 1))  # Legend outside the plot area
plt.grid(True)
plt.tight_layout()

# Show the plot
plt.show()
```

```python
# Now I want to create variables to pull limited countries into a plot

# All of the countries in this variable were identified by ACLED as having "conflicts to watch" in 2021.  Noticeably missing was Russia and Ukraine, who have been added at my discression.
# First #'ed code shows all of the countries potentially considered.  Actual is below.
# selected_countries = ['Ethiopia', 'India', 'Pakistan', 'Myanmar', 'Haiti', 'Belarus', 'Colombia', 'Armenia', 'Azerbaijan', 'Yemen', 'Mozambique', 'Mali', 'Niger', 'Burkina Faso', 'Russia', 'Ukraine']
selected_countries = ['Ethiopia', 'Pakistan', 'Myanmar', 'Haiti', 'Belarus', 'Colombia', 'Armenia', 'Azerbaijan', 'Yemen', 'Mozambique', 'Mali', 'Niger', 'Burkina Faso', 'Ukraine']

selected_df = df[df['Country'].isin(selected_countries)]

# Plot
plt.figure(figsize=(12, 8))
for country in selected_countries:
    country_data = selected_df[selected_df['Country'] == country]
    plt.plot(country_data['Year'], country_data['Spending (2020 USD)'], label=country)

# Set labels and title
plt.xlabel('Year')
plt.ylabel("Spending '000 (2020 USD)")
plt.title('Military Expenditure for Selected Countries 1988 to 2021')
plt.legend(loc='upper left', bbox_to_anchor=(1, 1))  # Legend outside the plot area
plt.grid(True)
plt.tight_layout()

# Show the plot
plt.show()

```

```python
# Now I want to create variables to pull limited countries into a plot

# Displayed here according to Percent of Government Spending
selected_countries = ['Ethiopia', 'Pakistan', 'Myanmar', 'Haiti', 'Belarus', 'Colombia', 'Armenia', 'Azerbaijan', 'Yemen', 'Mozambique', 'Mali', 'Niger', 'Burkina Faso', 'Ukraine']

selected_df = df[df['Country'].isin(selected_countries)]

# Plot
plt.figure(figsize=(12, 8))
for country in selected_countries:
    country_data = selected_df[selected_df['Country'] == country]
    plt.plot(country_data['Year'], country_data['Percent of Government Spending'], label=country)

# Set labels and title
plt.xlabel('Year')
plt.ylabel("Percent of Government Spending")
plt.title('Military Expenditure as Percent of Government Spending 1988 to 2021')
plt.legend(loc='upper left', bbox_to_anchor=(1, 1))  # Legend outside the plot area
plt.grid(True)
plt.tight_layout()

# Show the plot
plt.show()
```
